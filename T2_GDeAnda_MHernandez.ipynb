### Ejercicio 1: 
Obtener máximos y mínimos absolutos, usando `sympy`, de las funciones dadas en los intervalos dados, y graficar la función en dicho intervalo señalando los puntos máximo y mínimo.

$f(x)=2x^3+4x^2+1$ en $[-1.5,\,0.5]$

# Código de solución estudiante 1 Gustavo de Anda
    


```python
import sympy as sym
sym.init_printing(use_latex="mathjax")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
def f(x):
    return ((2 * x ** 3)+ (4 * x ** 2) + (1))
f(x)
```

```python
sym.var("x", real=True)
```

```python
df = sym.diff(f(x),x)
df
```

```python
x_c = sym.solve(df,x)
x_c
```

```python
f(-1.5),f(.5),f(-4/3),f(0)
```

```python
# Parte de la gráfica
xnum = np.linspace(-1.5,.5,150)
ynum = f(xnum)

# Tamaño de la gráfica
plt.figure(figsize=(12,8))

# Agregamos los gráficos
plt.plot(xnum,ynum,'--c',lw=6,label='$y=2x^3 + 4x^2 + 1$')
plt.plot([-4/3],[3.37037037037037],'Hb',ms=18,label='Punto máximo $(-4/3,3.37)$ en el intervalo')
plt.plot([0],[1],'or',ms=18,label='Punto mínimo $(0,1)$ en el intervalo ')

# Ejes de la gráfica
plt.xlabel('Eje de las x',fontsize=18)
plt.ylabel('Eje de las y',fontsize=18)

plt.grid(color='black')
plt.legend(loc='best')
plt.show()
```

### Respuesta planteada en el ejercicio usando 
Solo nos interesan los puntos cuando x vale -4/3 que es el punto máximo $(\frac{-4}{3},3.37)$ del intervalo y cuando x vale 0 que es el punto mínimo $(0,1)$ del intervalo $[-1.5,0.5]$

### Ejercicio 2:
Obtener máximos y mínimos absolutos, usando `sympy`, de las funciones dadas en los intervalos dados, y graficar la función en dicho intervalo señalando los puntos máximo y mínimo.

$g(x) = 5x^3-3x+10$ en el intervalo $[-1, \,0]$

# Código de solución estudiante 2 Maximiliano Hernández




```python
def g(x):
    return (5*x**3 - 3*x + 10)
g(x)
```

```python
sym.var('x', real=True)
```

```python
df = sym.diff(g(x),x)
df
```

```python
xc = sym.solve(df, x)
xc
```

```python
g(-(5**(1/2)/5)), g(-1), g(0), g((5**(1/2)/5))
```

```python
xnum= np.linspace(-1,0,254)
ynum = g(xnum)

plt.figure(figsize=(12,8))

plt.plot(xnum, ynum, '--c', lw=6, label= '$y=5x^3-3x+10$' )
plt.plot([-((5**(1/2)/5))], [10.8944271909999],'Hb', ms=18, label='Punto máximo $(-5^{(1/2}/5,10.894)$ en el intervalo')
plt.plot([-1],[8],'or', ms=18, label='Punto minimo $(-1,8)$ en el intervalo')

plt.xlabel('Eje de las x', fontsize=18)
plt.ylabel('Eje de las y', fontsize=18)

plt.grid()
plt.legend(loc='best')
plt.show()
```

### Respuesta planteada en el ejercicio usando 
Al evaluar la función en los extremos del intervalo y en los puntos cuando se deriva y se iguala a 0 la función obtuve como respuesta que el mínimo absoluto es en el punto (-1,8) y el maximo absoluto está ubicado en la coordenada $(-5^{(1/2)}/5,10.894)$
